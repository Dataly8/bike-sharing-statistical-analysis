 **El siguiente conjunto de datos (day.csv) contiene información acerca del uso de un servicio de alquiler de bicicletas.
 Para el año 2012, calcula la media de uso mensual (mes a mes de la columna “cnt”) e indica los 3 meses donde hubo más dispersión en el uso.**



```python
import pandas as pd
```

```python
public_csv_url ='https://raw.githubusercontent.com/Dataly8/bike_sharing_estadistica/main/day.csv'
```

```python
df = pd.read_csv(public_csv_url)
df
```

Filtro por el año 2012 que se representa con un 1 en la columna 'yr'.
(yr: 0 = 2011, 1 = 2012)

```python
df[df['yr'] == 1]
```

Agrupo por 'mnth' y promediamos la columna 'cnt'

```python
df[df['yr'] == 1].groupby('mnth')['cnt'].mean().round(2)
```

Ejecuto la desviación estandar con .std() que por defecto siempre hace el cálculo de la muestral (divide por n-1).

```python
df[df['yr'] == 1].groupby('mnth')['cnt'].std().round(2)
```

Cálculo del coeficiente de variación para entender la variación relativa de cada mes y comparar con la desviación estándar.

```python
media = df[df['yr'] == 1].groupby('mnth')['cnt'].mean().round(2)
std = df[df['yr'] == 1].groupby('mnth')['cnt'].std().round(2)
cv = std/media
cv.round(4)
```

Lo junto en una tabla comparativa:

```python
tabla = pd.DataFrame({'Media': media, 'Desv.Estándar': std, 'Coef.Variación' : cv})
```

```python
tabla.round(2)
```

### **Cálculo de la media mes a mes**:

- El uso medio de alquiler de bicicletas **crece en los meses más cálidos, y presenta un pico en el mes de septiembre con 7285.77 bicis al día**.

- La media **desciende considerablemente en diciembre y enero**.

Estos datos son coherente dado que se trata de un servicio al aire libre, así que lo esperado es que el alquiler y uso suban en los meses más soleados.

### **Medidas de dispersión:**

Para entender la dispersión de los datos, se han calculado dos medidas complementarias:



La **desviación estándar** (σ) que expresa la dispersión en unidades absolutas.

  *Meses con mayor ranking:*

    1.   Octubre (1941.82)
    2.   Diciembre (1803.23)
    3.   Abril (1308.94)


El **coeficiente de variación** que expresa expresa la dispersión en relación al nivel medio de uso de cada mes.

  *Meses con mayor ranking:*

    1.   Diciembre (45%)
    2.   Octubre (30%)
    3.   Enero (28%)


**Interpretación de los resultados:**

Octubre y Diciembre son los meses que aparecen con mayor dispersión, tanto en desviación estándar como en coefiente de correlación. La diferencia está en el tercer puesto: Abril entra en el ranking de desv. estándar, en cambio, Enero entra por coef. de correlación.

Esto se puede explicar porque la medida de coeficiente de correlación normaliza por el nivel de uso: Diciembre y Enero son meses de baja demanda y su variación es muy grande en proporción a su media.

Se puede establecer una hipótesis: los meses de mayor dispersión coinciden con periodos de transición estacional (en otoño y en invierno).
La inestabilidad del clima podría generar días de uso muy dispares.
